a school of fish can be used as an example of randomwalkers.In a school each individual fish moves in response to the neighboring fish. No fish makes a decision to move on its own and thus the byproduct of the group's overall movement is random. The school would move to gether as a whole and the direction of such is random. WE can also introduce a second school to act as a predator to the first. The existence of this fish will cause the original group to change systems to represent fleeing movement from the predators.

if no predator:
$(x,y,z) = 0.75f(x-\Delta x,y,z,t)+0.25f(x+\Delta x,y,z,t)+0.5(x,y-\Delta y,z,t)+0.5(x,y+\Delta y,z,t)+0.5(x,y,z-\Delta z,t)+0.5(x,y,z+\Delta z,t)$
if predator:
$(x,y,z) = 0.25f(x-\Delta x,y,z,t)+0.75f(x+\Delta x,y,z,t)+0.5(x,y-\Delta y,z,t)+0.5(x,y+\Delta y,z,t)+0.5(x,y,z-\Delta z,t)+0.5(x,y,z+\Delta z,t)$

The model fails to take into account the behavior of the fish on an individual scale and thus fails to cover the possibility of the school dispersing for cases such as threat of the predator or size of school being too large. This model assumes the school moves as a single unit in all cases. The parameters used are a good approximate for the movement of an actual school of fish.

```python
# base code borrowed from towardsdatascience.com
%pylab inline
from itertools import cycle
from mpl_toolkits.mplot3d import Axes3D
#colors = cycle(bgrcmykbgrcmykbgrcmykbgrcmyk)
# Define parameters for the walk
dims = 3
step_n = 1000
step_set = [-1, 0, 1]
origin = np.zeros((1,dims))
# Simulate steps in 3D
step_shape = (step_n,dims)
######Our Code######
def move():
    x=np.random.rand()
    y=np.random.rand()
    z=np.random.rand()
    if(x < 0.25):
        x=-1
    else:
        x=1
    if(y < 0.5):
        y=1
    else:
        y=-1
    if(z < 0.5):
        z=1
    else:
        z=-1
    return np.array([x,y,z])   
i=0
steps=np.zeros((step_n,3),int)
for i in range(step_n):
    temp=move()
    steps[i,0]=temp[0]
    steps[i,1]=temp[1]
    steps[i,2]=temp[2]
    i=i+1



#steps = np.random.choice(a=step_set, size=step_shape)
#steps=np.array([[x,y,z]])
path = np.concatenate([origin, steps]).cumsum(0)
########
start = path[:1]
stop = path[-1:]
# Plot the path
fig = plt.figure(figsize=(10,10),dpi=200)
ax = fig.add_subplot(111, projection='3d')
ax.grid(False)
ax.xaxis.pane.fill = ax.yaxis.pane.fill = ax.zaxis.pane.fill = False
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.scatter3D(path[:,0], path[:,1], path[:,2], 
             c='blue', alpha=0.25,s=1)
ax.plot3D(path[:,0], path[:,1], path[:,2], 
          c='blue', alpha=0.5, lw=0.5)
ax.plot3D(start[:,0], start[:,1], start[:,2], 
          c='red', marker='+')
ax.plot3D(stop[:,0], stop[:,1], stop[:,2], 
          c='black', marker='o')
plt.title('no predator')
#plt.savefig('plots/random_walk_3d.png',dpi=250);
```
