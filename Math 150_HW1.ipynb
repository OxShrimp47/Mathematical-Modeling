Reynolds number is a number used is predicting flow pattererns of liquids. It is defined by a multitude of parameters that each are each defined by their own set of dimensions; but yet Reynolds number lacks any such dimensions.Reynolds number is defined as:

$ Re= \frac{\rho u L}{ \mu}=\frac{u L}{v}$

Substituting these parameters for their dimensions shows why Reynolds number has no deimension.

$\rho = kg/m^3$

$u = m/s$

$L= m$

$\mu = Ns/m = ((kg *m)/s)/m^2=kg/(m *s)$

$v = m^2/s$

$ Re =  \frac{\rho* u* L}{ \mu}= \frac{(kg/m^3) (m/s)(m)}{ (kg/(m* s))}= \frac{kg/(m *s)}{ kg/(m *s)}$

$ Re =\frac{u *L}{v}= \frac{(m/s)(m)}{ (m^2/s)}= \frac{m^2/s}{ m^2/s}$
